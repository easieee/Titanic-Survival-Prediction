```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
ts_train = pd.read_csv('train.csv')
ts_test = pd.read_csv('test.csv')
tsdata = pd.concat([ts_train, ts_test])
tsdata = tsdata.reset_index(drop=True)
display(tsdata)
tsdata.isnull().sum()
```

```python
tsdata = tsdata.drop(["PassengerId", "Pclass", "Name",
                     'Age', 'Parch', 'Ticket', 'Fare',
                     'Cabin','Embarked'], axis=1)
tsdata
```

```python
cl_tsdata = tsdata['Survived'].isnull()==False 
trn_tsdata = tsdata[cl_tsdata]
trn_tsdata['Survived'] = trn_tsdata['Survived'].astype(int)

trn_tsdata
```

```python
trn_tsdata['Sex'].value_counts()
```

```python
X = trn_tsdata.drop(columns = ['Sex','SibSp'], axis=1)
X.head()
```

```python
y = trn_tsdata['Survived']
y.head()
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_std = std.fit_transform(X)
X_std
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.20, random_state=0)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)
logr.fit(X_train, y_train)
logr.score(X_test, y_test)
```
